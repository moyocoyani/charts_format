## Annotations.


Most of the time we used to share charts without further information and this is a practice that should be avoided at any cost.

Accordingly to Andy Kirk (*Kirk, Andy. Data visualisation: A handbook for data driven design. Sage, 2016.*) we could have several types of annotations in our visualization. From headings, introduction and footnotes, to axis titles, axis labels and value labels. So let's focus on headings, axis title, axis labels and value labels and how to create those elements with matplotlib.

`ax.set_xlabel(...)`

This title property (`suptitle`) is beyond the axes space ([check the matplotlib object hierarchy](https://files.realpython.com/media/fig_map.bc8c7cabd823.png)), and we can access to it by invoking the pyplot function (`plt`) or calling our figure element (`fig`), i.e. `plt.suptitle` or `fig.suptitle`.

# 1. Text alignment.

One best practice is to align the title and subtitle to the left, because [center-alignment looks a little messy](http://www.storytellingwithdata.com/blog/2012/09/quick-tip-left-uppermost-align-title).

With matplotlib we can do that by modifying the "location" of the axes title, so let's move the location to the *left* with `loc=left`



```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt)
ax.set_ylabel(label_yt)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t)

#displaying data
plt.show()
```

and for the suptitle we need to do something tricky.

First, a property called `horizontalalignment` needs to be setted to the`left`

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt)
ax.set_ylabel(label_yt)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left')

#displaying data
plt.show()
```

As you can see, the text was not left aligned as we expected. Well, matplotlib uses a coordinate system in order to indicate where to place some text and `horizontalalignment` tells to matplotlib which part of the text will be attached to that coordinate.

In default settings, matplotlib indicates that `suptitle` must be in the top-middle of the chart and the center of the text is the anchor point that is attached to the mid section. Hence, if we change the horizontal alignment, without modifying the coordinates, we are only indicating to matplotlib that the anchor point is the leftmost part of the text, and this is the starting point of the title.

In order to move our text to the left uppermost section of the chart, we need to specify the coordinates but we need to specify that our coordinate system corresponds to the figure and not the data points.

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt)
ax.set_ylabel(label_yt)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0,
             y=1,
             transform=fig.transFigure)

#displaying data
plt.show()
```

The coordinates $x=0$,$y=1$, in the figure coordinate system, indicates to matplotlib that we want to place an element at the very beginning of the canvas. Remember, the $x$ coordinates goes from $0$ (left) to $1$ (right) where 0 is the leftmost part, and $y$ goes from $0$ (bottom) to $1$ (top) where $1$ is the uppermost part.

Now that we understand the coordinate system, what we want is to align the title and subtitle with our y-axis label, and this is around $x\approx0.05$ in figure-coordinate system with a `width=11`; if you change the width of the plot, you will need to adjust the value of $x$.

Let's move `plt.suptitle` to ($x=0.05$,$y=1$) and `ax.title` to ($x=0.05$,$y=0.9$) in the figure coordinates 

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt)
ax.set_ylabel(label_yt)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure)

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure)

#displaying data
plt.show()
```

The subtitle is close to the chart, we can give it an extra space by modifying the parameter `subplots_adjust` from `top=90` to `top=0.82` (`plt.subplots_adjust(top=0.82)`)

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt)
ax.set_ylabel(label_yt)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure)

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure)

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

Continuing with the recommendations from the blog [storytelling with data](http://www.storytellingwithdata.com/blog/2012/09/quick-tip-left-uppermost-align-title), we can align the *x-axis title* to the leftmost and the *y-axis title* to the uppermost (or leave it unaltered, is your choice). 

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left")
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right")
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)
ax.tick_params(axis='both')

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure)

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure)

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

Now we are ready to give some format to the text.

# 2. Font size.

The default size of our text is hard to read and remember, we want to help our viewers to consume the visualization. 

Let's size up the text elements in the chart.

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left",
              fontsize=16)
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right",
              fontsize=16)
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)

ax.tick_params(axis='both',
              labelsize=16)

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure,
            fontsize=16)

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t,
        fontsize=16)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure,
             fontsize=16)

plt.subplots_adjust(top=0.82)


#displaying data
plt.show()
```

it is easier to read, isn't it? Well now we can apply our first contrast technique.

All text elements look pretty the same, so we need to create a contrast in order to distinguish headings from axis labels or value labels, and the fontsize is one element that we can use in order to create that contrast.

Let's size by the next parameters:
- Title: 24pts
- Subtitle: 20pts
- Axis titles: 16pts
- Axis labels: 12 pts
- Value labels: 12pts 

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left",
              fontsize=16)
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right",
              fontsize=16)
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)

ax.tick_params(axis='both',
              labelsize=12)

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure,
             fontsize=20)

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t,
        fontsize=12)

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure,
             fontsize=24)

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

# 3. Fontweight.

With a proper fontweight we can add contrast to the text elements by setting the thickness of the characters in a text. The most common options for selecting the weight in matplotlib are *light* , *regular* and *bold* and it depends on the font. Some typefaces in matplotlib only have one or two weights (regular and bold are the most common weights) but they are enough to create a contrast.

Let's add some weight to our text:
- Title: 24pts; bold
- Subtitle: 20pts; regular
- Axis titles: 16pts; regular
- Axis labels: 12 pts; regular
- Value labels: 12pts; regular

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left",
              fontsize=16,
              fontweight="regular")
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right",
              fontsize=16,
              fontweight="regular")
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)

ax.tick_params(axis='both',
              labelsize=12)

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure,
             fontsize=20,
             fontweight="regular")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t,
        fontsize=12,
        fontweight="regular")

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure,
             fontsize=24,
             fontweight="bold")

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

# 4. Pairing fonts.

There is a font classification based on families:
- serif.
- sans serif.
- handwriting.
- monospace.
- fantasy or display.

Depending on the audience and context you need to perform an editorial decision to select the most adequate font family. However, *serif* and *sans serif* are the most recommended options.

In previous examples, the font family has been sans serif so far and we can check it by typing the property `fontfamily="sans"` in our text elements.

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left",
              fontsize=16,
              fontweight="regular",
              fontfamily="sans")
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right",
              fontsize=16,
              fontweight="regular",
              fontfamily="sans")
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)

ax.tick_params(axis='both',
              labelsize=12)

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure,
             fontsize=20,
             fontweight="regular",
             fontfamily="sans")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t,
        fontsize=12,
        fontweight="regular",
        fontfamily="sans")

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure,
             fontsize=24,
             fontweight="bold",
             fontfamily="sans")

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

As you can tell there is no difference with the current chart and the previous format, however, if we change the `fontfamily` parameter to `serif` (in `plt.suptitle` and `ax.title`) we obtain the following chart: 

```
#canvas: Keep a 16:9 aspect ratio with 100dpi 
fig,ax=plt.subplots(1,1,figsize=(11,6.9),dpi=100)

#******DATA
#ploting data
ax.plot(xtime,ytime)

#******TEXT
#axis labels
ax.set_xlabel(label_xt,
              ha="left",
              fontsize=16,
              fontweight="regular",
              fontfamily="sans")
ax.xaxis.set_label_coords(x=0, y=-0.07, transform=ax.transAxes)

ax.set_ylabel(label_yt,
              ha="right",
              fontsize=16,
              fontweight="regular",
              fontfamily="sans")
ax.yaxis.set_label_coords(x=-0.07, y=1, transform=ax.transAxes)

ax.tick_params(axis='both',
              labelsize=12)

#subtitle
ax.set_title(subtitle_t,
             loc="left",
             x=0.05,
             y=0.9,
             transform=fig.transFigure,
             fontsize=20,
             fontweight="regular",
             fontfamily="serif")

#chart_context
ax.text(x=xtime[np.argmax(ytime)],
        y=ytime.max(),
        s=context_t,
        fontsize=12,
        fontweight="regular",
        fontfamily="sans")

#superior title
plt.suptitle(title_t,
             horizontalalignment = 'left',
             x=0.05,
             y=1,
             transform=fig.transFigure,
             fontsize=24,
             fontweight="bold",
             fontfamily="serif")

plt.subplots_adjust(top=0.82)

#displaying data
plt.show()
```

And now our text format is summarized by

- Title: serif; 24pts; bold
- Subtitle: serif; 20pts; regular
- Axis titles: sans serif; 16pts; regular
- Axis labels: sans serif; 12 pts; regular
- Value labels: sans serif; 12pts; regular

Pairing serif and sans serif are one of the most common pairing techniques and this is what we have done so far. But there are many ways to pair typefaces. # 5. Select your font.

If you want to know what fonts do you have installed in matplotlib but also want to see a printed example of those fonts you can write the following code from [Jonathan Soma (List all fonts available in matplotlib plus samples)](http://jonathansoma.com/lede/data-studio/matplotlib/list-all-fonts-available-in-matplotlib-plus-samples/
)
